# Information

### Event_msg_type documentation:
    1   -->   Made Shot              (add points, possession ends & changes)
    2   -->   Missed Shot            (don't add points, don't change possession --> DONT CARE)
    3   -->   Free Throw             (add points, possession remain same)
    4   -->   Rebound                (possesion could change, remain same)
    5   -->   Turnover               (possession changes)
    6   -->   Foul                   (don't add points, don't change possession) (what about offensive foul though?? does a turnover event follow?)
    7   -->   Violation              (what about offensive violation though?? does a turnover event follow?)
    8   -->   Substitution           (change active players)
    9   -->   Timeout                (DONT CARE)
    10  -->   Jumpball               (gotta figure out possession after jumpball)
    11  -->   Ejection               (DONT CARE)
    12  -->   Start Period
    13  -->   End Period             (possession ends)
    14  -->   Validation             (DIDNT EXIST)
    15  -->   Memo                   (DIDNT EXIST)
    16  -->   Final Box Printed      (DONT CARE)
    18  -->   Instant Replay         (DONT CARE)
    20  -->   Stoppage               (DONT CARE)

# Define Classes

```python
class Game():
    """
    Records all the information of one game
    """
    def __init__(self, game_id, df_play_by_play, df_game_lineup):
        self.game_id = game_id
        self.df_play_by_play = df_play_by_play.copy().reset_index(drop=True)
        self.df_game_lineup = df_game_lineup.copy().reset_index(drop=True)
        
        # get the 2 teams of this game
        self.team_A_id, self.team_B_id = self.df_game_lineup['Team_id'].unique().tolist()
        
        # get the players of the 2 teams
        self.team_A_players = self.df_game_lineup[self.df_game_lineup['Team_id'] == self.team_A_id]['Person_id'].unique().tolist()
        self.team_B_players = self.df_game_lineup[self.df_game_lineup['Team_id'] == self.team_B_id]['Person_id'].unique().tolist()
        
        # create 2 Team objects
        self.team_A = Team(self.game_id, self.team_A_id, self.team_A_players)
        self.team_B = Team(self.game_id, self.team_B_id, self.team_B_players)
        
        # sort the play_by_play data
        self.df_play_by_play.sort_values(by=['Period','PC_Time','WC_Time','Event_Num'], ascending=[True,False,True,True], inplace=True)
        
        # keep track of possession status
        self.team_on_offense = ""
        self.team_on_defense = ""
        self.team_start_period_possession = ""             #keeps track of team that started period off poss.
        # keep track of free throw progress
        self.free_throw_happening_time = -1
        self.tech_free_throw_happening_time = -1
        self.flagrant_free_throw_happening_time = -1
        self.clear_path_free_throw_happening_time = -1
        
        # access team object easier with team_id
        self.team_map = { self.team_A_id : self.team_A, self.team_B_id : self.team_B }
        
        
        
    def __str__(self):
        ret = "Game ID:\t{}\n".format(self.game_id)
        ret += "\n\tTeam A ID:\t{}\n".format(self.team_A_id)
        ret += "\tTeam B ID:\t{}\n\n".format(self.team_B_id)
        return ret
    
    def check_team(self, player_id):
        """
        Returns team_id of team which player belongs to
        """
        if player_id in self.team_A_players:
            return self.team_A_id
        else:
            return self.team_B_id
    
    def switch_team(self, team_id):
        """
        Returns the other team's team_id
        """
        if team_id == self.team_A_id:
            return self.team_B_id
        else:
            return self.team_A_id
        
    
    
    def start_of_period(self, period):
        # set active players for both teams at start of quarter
        team_A_active_players = self.df_game_lineup[(self.df_game_lineup['Period'] == period) & 
                                                    (self.df_game_lineup['Team_id'] == self.team_A_id)
                                                   ]['Person_id'].unique().tolist()
        team_B_active_players = self.df_game_lineup[(self.df_game_lineup['Period'] == period) & 
                                                    (self.df_game_lineup['Team_id'] == self.team_B_id)
                                                   ]['Person_id'].unique().tolist()
        self.team_A.set_active_players(team_A_active_players)
        self.team_B.set_active_players(team_B_active_players)
        
    
    def process_game(self):
        """
        Goes through all the play_by_play data for one game
        """
        
        period = 0
        for idx, row in self.df_play_by_play.iterrows():
            # if new period, do start of period logistics
            if (row['Period'] != period):
                period = row['Period']
                self.start_of_period(period)
                
                # record which team gets start of period off-possession
                if period == 1 or period > 4:
                    next_row = self.df_play_by_play.iloc[idx+1]
                    if (next_row['Event_Msg_Type'] != 10):
                        print("ERROR! First event of period {} is not jumpball!".format(period))
                    else:
                        self.team_start_period_possession = self.check_team(next_row['Person1'])
                        # jumpball event will be processed and determine possession
                # periods 2,3,4 just exchange start of period possessions
                else:
                    self.team_start_period_possession = self.switch_team(self.team_start_period_possession)
                    # possession is automatically awarded to correct team
                    self.team_on_offense = self.team_start_period_possession
                    self.team_on_defense = self.switch_team(self.team_on_offense)
            
            # process play_by_play data row by row
            self.process_one_row(row)
            
            
            
                
                
    def process_one_row(self, row):
        """
        Processes one row of play_by_play data
        """
        event_msg = row['Event_Msg_Type']
        action_type = row['Action_Type']
        person_1 = row['Person1']
        person_2 = row['Person2']
        option_1 = row['Option1']
        person_1_team = self.check_team(person_1)
        person_2_team = self.check_team(person_2)
        pc_time = row['PC_Time']
        period = row['Period']
        
        
        # check normal free throws end. if so, move active_2 into active_1, clear active_2
        if (self.free_throw_happening_time != -1):
            # clock has moved, free throws have ended
            if (pc_time != self.free_throw_happening_time):
                self.free_throw_happening_time = -1
                self.team_map[self.team_A_id].move_active_2_into_active_1()
                self.team_map[self.team_B_id].move_active_2_into_active_1()
                
        if (self.tech_free_throw_happening_time != -1):
            # clock has moved, free throws have ended
            if (pc_time != self.tech_free_throw_happening_time):
                self.tech_free_throw_happening_time = -1
                
                #move active_2 into active_1 only when tech free throws are not happening in the middle of other FTs
                if (self.free_throw_happening_time == -1 and
                    self.flagrant_free_throw_happening_time == -1 and
                    self.clear_path_free_throw_happening_time == -1):
                    self.team_map[self.team_A_id].move_active_2_into_active_1()
                    self.team_map[self.team_B_id].move_active_2_into_active_1()
        
        if (self.flagrant_free_throw_happening_time != -1):
            # clock has moved, free throws have ended
            if (pc_time != self.flagrant_free_throw_happening_time):
                self.flagrant_free_throw_happening_time = -1
                self.team_map[self.team_A_id].move_active_2_into_active_1()
                self.team_map[self.team_B_id].move_active_2_into_active_1()
                
        if (self.clear_path_free_throw_happening_time != -1):
            # clock has moved, free throws have ended
            if (pc_time != self.clear_path_free_throw_happening_time):
                self.clear_path_free_throw_happening_time = -1
                self.team_map[self.team_A_id].move_active_2_into_active_1()
                self.team_map[self.team_B_id].move_active_2_into_active_1()
        
        
        
        
        if event_msg == 12:          # start of period
            pass
        if event_msg == 2:           # missed shot
            pass
        if event_msg == 9:           # time-out
            pass
        if event_msg in (11,14,15,16,18,20):       # other events
            pass
                
                
        
        if event_msg == 1:           # made shot
            self.add_points(option_1)
            self.end_possession()
            self.switch_possession()
        
        
        if event_msg == 3:           # free throw
            # if it is a tech FT (no possession awarded), we don't count additional possession
            if action_type == 16:
                if option_1 == 1:
                    self.add_points(option_1)
            
            # if it is not the last attempt of FT, just count point if made 
            if action_type in (11, 13, 18, 25):
                if option_1 == 1:
                    self.add_points(option_1)
            
            # If it is the last attempt of clear path/flagrant FT 
            if action_type in (19, 20, 26):
                if option_1 == 1:
                    self.add_points(option_1)
                # we add an additional possession
                self.end_possession()
                # if there is possession switch, we switch
                if self.check_team(person_1) == self.team_on_defense:
                    self.switch_possession()
            
            # If it is the last attempt of normal FT
            if action_type in (10, 12, 15):
                if option_1 == 1:
                    self.add_points(option_1)
                    self.end_possession()
                    self.switch_possession()
                else:
                    # The FT didn't go in and there's rebound after this, so we don't need to worry now
                    pass
        
        if event_msg == 4:           # rebound
            team_got_rebound = self.check_team(person_1) 
            if team_got_rebound == self.team_on_defense:
                self.end_possession()
                self.switch_possession()
        
        if event_msg == 5:           # turnover
            self.end_possession()
            self.switch_possession()
        
        if event_msg == 6:           # foul
            # double foul or double technical foul
            if action_type in (10, 16):
                # do nothing, no free throws will be awarded, possession continues
                pass
            
            # offensive foul:
            if action_type in (4, 26):
                # do nothing, turnover event always follows offensive foul
                pass
            
            # record game-time of foul happening
            if action_type in (1,2,3,5,6,27,28,29,4,26):            # normal foul
                self.free_throw_happening_time = pc_time
            if action_type in (11, 17):                                # technical foul
                self.tech_free_throw_happening_time = pc_time
            if action_type in (14,15):                                # flagrant foul
                self.flagrant_free_throw_happening_time = pc_time
            if action_type == 9:                                     #clear path foul
                self.clear_path_free_throw_happening_time = pc_time
        
        
        if event_msg == 7:           # violation
            if action_type in (3,5):
                team_perform_violation = self.check_team(person_1)
                if team_perform_violation == self.team_on_offense:
                    self.end_possession()
                    self.switch_possession()
        
        if event_msg == 8:           # substitution
            #person_1 is leaving game, person_2 is entering game
            
            team_id = self.check_team(person_1)
            before_sub_active_players = self.team_map[team_id].get_active_players()
            after_sub_active_players = before_sub_active_players.copy()
            after_sub_active_players.remove(person_1)
            after_sub_active_players.append(person_2)
            
            # if free throws are happening, record roster changes into active_2
            if (self.free_throw_happening_time != -1 or
                self.tech_free_throw_happening_time != -1 or
                self.flagrant_free_throw_happening_time != -1 or
                self.clear_path_free_throw_happening_time != -1):
                
                self.team_map[team_id].set_active_players(active_player_ids = after_sub_active_players,
                                                          active_2 = True)
                
            else:
                self.team_map[team_id].set_active_players(active_player_ids = after_sub_active_players,
                                                          active_2 = False)
        
        if event_msg == 10:          # jump ball
            # if it is start of first quarter or overtime the possession belongs to
            # team where person_1 is on
            if (period == 1 and pc_time == 7200) or (period > 4 and pc_time == 3000):
                self.team_on_offense = self.check_team(person_1)
                self.team_on_defense = self.switch_team(self.team_on_offense)
            else:
                # if the defense gets the ball we start a new possession and switch the possession
                if self.check_team(person_1) == self.team_on_defense:
                    self.end_possession()
                    self.switch_possession()
                # Else the offense keeps going. The idea is that for jump ball if the offense gets the ball
                # the clock does not reset, so it makes more sense to treat it as a continuous possession
                else:
                    pass
        
        if event_msg == 13:          # end of period
            self.end_possession()
         
            
    
    def end_possession(self):
        """
        End offensive possession for off team
        End defensive possession for def team
        """
        if (self.tech_in_other_foul_special_case()):
            self.team_map[self.team_on_offense].add_off_possessions(1, True)
            self.team_map[self.team_on_defense].add_def_possessions(1, True)
        else:
            self.team_map[self.team_on_offense].add_off_possessions(1)
            self.team_map[self.team_on_defense].add_def_possessions(1)
        
        
    
    def switch_possession(self):
        """
        Switch the teams on offensive and defensive possession
        """
        if self.team_on_offense == self.team_A_id:
            self.team_on_offense = self.team_B_id
            self.team_on_defense = self.team_A_id
        else:
            self.team_on_offense = self.team_A_id
            self.team_on_defense = self.team_B_id
        
    
    def add_points(self, points):
        """
        Add points scored for off team
        Add points allowed for def team
        """
        if (self.tech_in_other_foul_special_case()):
            self.team_map[self.team_on_offense].add_points_scored(points, True)
            self.team_map[self.team_on_defense].add_points_allowed(points, True)
        else:
            self.team_map[self.team_on_offense].add_points_scored(points)
            self.team_map[self.team_on_defense].add_points_allowed(points)
    
            
    def end_game(self):
        """
        Returns a pandas dataframe of final results 
        game_id, player_id, off_rtg, def_rtg
        """
        
        results = pd.DataFrame(columns = ['Game_ID', 'Player_ID', 'OffRtg', 'DefRtg'])
        
        team_A_roster = self.team_map[self.team_A_id].roster
        team_B_roster = self.team_map[self.team_B_id].roster
        game_id = self.game_id
        
        for player_id, player in team_A_roster.items():
            off_rtg = player.get_offensive_rating()
            def_rtg = player.get_defensive_rating()
            
            results = results.append({'Game_ID' : game_id,
                                      'Player_ID': player_id,
                                      'OffRtg': off_rtg,
                                      'DefRtg': def_rtg}, ignore_index=True)
        for player_id, player in team_B_roster.items():
            off_rtg = player.get_offensive_rating()
            def_rtg = player.get_defensive_rating()
            
            results = results.append({'Game_ID' : game_id,
                                      'Player_ID': player_id,
                                      'OffRtg': off_rtg,
                                      'DefRtg': def_rtg}, ignore_index=True)
        return results
    
    def tech_in_other_foul_special_case(self):
        # returns True if a technical foul is happening during another type of foul, false otherwise
        if ((self.tech_free_throw_happening_time != -1 and self.free_throw_happening_time != -1) or
            (self.tech_free_throw_happening_time != -1 and self.flagrant_free_throw_happening_time != -1) or
            (self.tech_free_throw_happening_time != -1 and self.clear_path_free_throw_happening_time != -1)):
            return True
        else:
            return False
    
```

```python
class Team():
    """
    Records all the information of one team in one game
    """
    def __init__(self, game_id, team_id, player_ids_list = []):
        self.game_id = game_id
        self.team_id = team_id
        
        # player_ids mapped to corresponding Player object
        self.roster = {}
        for player_id in player_ids_list:
            self.roster[player_id] = Player(self.game_id, self.team_id, player_id)
        
        # need 2 active lists to keep track of active players during fouls/free throws
        # if active_2 exists, active_2 contains most up-to-date players on court
        # if active_2 is empty, active_1 contains most up-to-date players on court
        self.active_1 = []
        self.active_2 = []
        
    
    def __str__(self):
        ret = "Team ID:\t{}\n\n".format(self.team_id)
        for idx, player_id in enumerate(self.roster):
            ret += "\tPlayer {}:\t{}\n".format(idx+1, player_id)
        
        return ret
    
    def move_active_2_into_active_1(self):
        if self.active_2:
            self.active_1 = self.active_2.copy()
            self.active_2 = []
    
    def set_active_players(self, active_player_ids = [], active_2 = False):
        if active_2:
            self.active_2 = active_player_ids.copy()
        else:
            self.active_1 = active_player_ids.copy()
    
    def get_active_players(self):
        # returns active_2 if active_2 not empty
        if not self.active_2:
            return self.active_1
        else:
            return self.active_2
    
    def get_player(self, player_id):
        return self.roster[player_id]
    
    def add_points_scored(self, points, tech_in_other_foul_special_case = False):
        if tech_in_other_foul_special_case:
            if self.active_2:
                for player_id in self.active_2: 
                    self.roster[player_id].add_points_scored(points)
            else:
                for player_id in self.active_1: 
                    self.roster[player_id].add_points_scored(points)
        else:
            for player_id in self.active_1: 
                self.roster[player_id].add_points_scored(points)
        
    def add_points_allowed(self, points, tech_in_other_foul_special_case = False):
        if tech_in_other_foul_special_case:
            if self.active_2:
                for player_id in self.active_2: 
                    self.roster[player_id].add_points_allowed(points)
            else:
                for player_id in self.active_1: 
                    self.roster[player_id].add_points_allowed(points)
        else:
            for player_id in self.active_1: 
                self.roster[player_id].add_points_allowed(points)
    
    def add_off_possessions(self, number, tech_in_other_foul_special_case = False):
        if tech_in_other_foul_special_case:
            if self.active_2:
                for player_id in self.active_2: 
                    self.roster[player_id].add_off_possessions(number)
            else:
                for player_id in self.active_1: 
                    self.roster[player_id].add_off_possessions(number)
        else:
            for player_id in self.active_1: 
                self.roster[player_id].add_off_possessions(number)
    
    def add_def_possessions(self, number, tech_in_other_foul_special_case = False):
        if tech_in_other_foul_special_case:
            if self.active_2:
                for player_id in self.active_2: 
                    self.roster[player_id].add_def_possessions(number)
            else:
                for player_id in self.active_1: 
                    self.roster[player_id].add_def_possessions(number)
        else:
            for player_id in self.active_1: 
                self.roster[player_id].add_def_possessions(number)
    
```

```python
class Player():
    def __init__(self, game_id, team_id, player_id):
        self.game_id = game_id
        self.team_id = team_id
        self.player_id = player_id
    
        self.points_scored = 0
        self.points_allowed = 0
        self.off_possessions = 0
        self.def_possessions = 0
    
    def __str__(self):
        ret = "Player ID: {}\n".format(self.player_id)
        ret += "\tPoints Scored: {}\n".format(self.points_scored)
        ret += "\tPoints Allowed: {}\n".format(self.points_allowed)
        ret += "\tOff Possessions: {}\n".format(self.off_possessions)
        ret += "\tDef Possessions: {}\n".format(self.def_possessions)
        ret += "\tOffensive Rating: {}\n".format(self.get_offensive_rating())
        ret += "\tDefensive Rating: {}\n".format(self.get_defensive_rating())
        
        return ret
    
    def add_points_scored(self, points):
        self.points_scored += points
        
    def add_points_allowed(self, points):
        self.points_allowed += points
    
    def add_off_possessions(self, number):
        self.off_possessions += number
    
    def add_def_possessions(self, number):
        self.def_possessions += number
    
    def get_offensive_rating(self):
        return self.points_scored / self.off_possessions * 100 if self.off_possessions != 0 else 0
    
    def get_defensive_rating(self):
        return self.points_allowed / self.def_possessions * 100 if self.def_possessions != 0 else 0
```
